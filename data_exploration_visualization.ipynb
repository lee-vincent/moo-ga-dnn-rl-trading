```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import mplfinance as mpf
import matplotlib.pyplot as plt

# Initial set up
%matplotlib inline
sns.set_palette("mako")
```

```python
# Gather data
norm_df = pd.read_csv("../data/processed/processed_data_tqqq_norm.csv")
prep_df = pd.read_csv("../data/processed/processed_data_tqqq.csv")
```

```python
norm_df.dtypes
```

```python
type(norm_df['date'][0])
```

```python
prep_df.set_index('date', inplace=True)
```

```python
# Make the date index compatible with mplfinance
prep_df.index = pd.to_datetime(prep_df.index)
```

```python
# Access the first 5 columns and the 7th column
prep_df.iloc[:, list(range(4)) + [5]]
```

```python
# Display plot
# mav = moving average (4 days, 8 days, 10 days)
mpf.plot(prep_df.iloc[:, list(range(4)) + [5]][-75:], type='candle', mav=(4, 8, 10), volume=True)
```

```python
# Revert the index to its original form
norm_df.reset_index(inplace=True)
```

All data points are float64 except for the date which is an object/string

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=norm_df, x='date', y='velocity_64w_0ts', label="Velocity")
sns.lineplot(data=norm_df, x='date', y='atr_64w_0ts', color="orange", label="ATR")
plt.title('Velocity and ATR Over Time')
plt.ylabel('Value')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=norm_df[-500:], x='date', y='velocity_64w_0ts', label='Velocity', palette="mako")
sns.lineplot(data=norm_df[-500:], x='date', y='atr_64w_0ts', color="orange", label='ATR')
plt.title('Velocity and ATR Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

## High, Low, Close, and Volume

```python
selected_columns = ['high', 'low', 'close', 'volume']
pairplot = sns.pairplot(norm_df[selected_columns], diag_kind='hist', height=2.5)

pairplot.map_lower(sns.regplot, line_kws={'color': 'orange'})
# pairplot.map_upper(sns.scatterplot)
for i, ax in enumerate(pairplot.diag_axes):
    sns.histplot(norm_df[selected_columns[i]], kde=True, color='orange', ax=ax, bins=3)

plt.show()
```

## 16 day Window with multiple time shifts (2, 6, 10)
